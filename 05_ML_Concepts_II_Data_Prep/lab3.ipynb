##### College Completion Dataset

Independent Business Metric: grad_150_value

##### Data Preparation

```python
import kagglehub
import os
import pandas as pd
import numpy as np 
import matplotlib
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import MinMaxScaler, StandardScaler

path = kagglehub.dataset_download("thedevastator/boost-student-success-with-college-completion-da")

dataset_path = "/home/vscode/.cache/kagglehub/datasets/thedevastator/boost-student-success-with-college-completion-da/versions/1"
print(os.listdir(dataset_path))
```

```python
file_name = "cc_institution_details.csv" 
df = pd.read_csv(os.path.join(dataset_path, file_name))

df.info()
```

```python
df2 = df[['level','control','student_count','awards_per_value','awards_per_state_value','awards_per_natl_value',
          'exp_award_value', 'exp_award_state_value', 'exp_award_natl_value', 'ft_pct','med_sat_value',
          'aid_value','grad_150_value']]
```

```python
df2 = df2.dropna()
```

```python
df2[['level','control']] = df2[['level','control']].astype('category')

print(df2.dtypes)
```

```python
abc = list(df2.select_dtypes('number'))

df2[abc] = MinMaxScaler().fit_transform(df2[abc])
df2
```

```python
category_list = list(df2.select_dtypes('category')) 
category_list
df3 = pd.get_dummies(df2, columns = category_list) 
df3
```

```python
df3 = df3.drop(['student_count'],axis=1)
```

```python
df3.grad_150_value.describe()
```

```python
df3['grad_150'] = pd.cut(df3.grad_150_value, bins = [-1,0.67,1], labels =[0,1])
prevalence = df3.grad_150.value_counts()[1]/len(df3.grad_150)
prevalence 
```

```python
Train, Test = train_test_split(df3,  train_size = .8, stratify = df3.grad_150) 
Tune, Test = train_test_split(Test,  train_size = .5, stratify= Test.grad_150)
print(Train.shape)
print(Test.shape)
print(Tune.shape)
```

##### Job Placement Dataset

##### Independent Business Metric: Salary

##### Data Preparation

```python
url = "https://raw.githubusercontent.com/DG1606/CMS-R-2020/master/Placement_Data_Full_Class.csv"

ds = pd.read_csv(url)

ds.head()
```

```python
ds=ds.dropna()
```

```python
ds.info()
```

```python
ds[['ssc_b','hsc_b','hsc_s','degree_t','workex','specialisation']] = ds[['ssc_b','hsc_b','hsc_s','degree_t','workex','specialisation']].astype('category')

print(ds.dtypes)
```

```python
abc = list(ds.select_dtypes('number'))

ds[abc] = MinMaxScaler().fit_transform(ds[abc])
ds
```

```python
category_list = list(ds.select_dtypes('category')) 
category_list
ds = pd.get_dummies(ds, columns = category_list) 
ds
```

```python
ds = ds.drop(['sl_no','gender','status'],axis=1)
```

```python
ds.salary.describe()
```

```python
ds['salaryf'] = pd.cut(ds.salary, bins = [-1,0.135,1], labels =[0,1])
prevalence = ds.salaryf.value_counts()[1]/len(ds.salaryf)
prevalence 
```

```python
Train, Test = train_test_split(ds,  train_size = .8, stratify = ds.salaryf) 
Tune, Test = train_test_split(Test,  train_size = .5, stratify= Test.salaryf)
print(Train.shape)
print(Test.shape)
print(Tune.shape)
```
