Question 1: How does a manufactuerer influence the Nutrition of the Cereal

Pseudocode: 
1. Group by manufacturer
2. What is the average calories
3. What is the average grams of sugar
4. Rank manufacturer by the calories 

```python
print(data.info()) 
```

```python
x1 = data.groupby(['mfr'])
```

Average of Calories, Sugar

```python
data2 = pd.DataFrame()
x2 = data2.assign(meanCalories = x1.calories.mean(), 
                    meanSugar=x1.sugars.mean())

x2
```

```python
x2.isnull().sum()
```

```python
x3 = x2.sort_values(by='meanCalories', ascending=False)
x3.head(10)
```

```python
def meancalories_manufacturer(x):
    x1 = x.groupby(['mfr'])
    data2 = pd.DataFrame()
    x2 = data2.assign(meanCalories = x1.calories.mean(), 
                    meanSugar=x1.sugars.mean()).sort_values(by='meanCalories', ascending=False)
    return x2
```

```python
print(meancalories_manufacturer(data))
```

```python
data = pd.read_csv('winequality-red-ddl.csv')
data.head()
```

Question 1: What is the quality of more acidic drinks? 

1. Limit ph scale to under 3.2
2. Display acidity, quality, text_rank
3. Calculate the total acidity of the drinks
4. Rank quality by total acidity 

```python
data.isnull().sum()
```

```python
data.dropna()
```

```python
data = data.query('pH <= 3.2')
```

```python
x1 = data.loc[:,['fixed acidity', 'volatile acidity','quality','text_rank']]
x1.head()
```

```python
x2 = x1.assign(total_acidity=data["fixed acidity"] + data["volatile acidity"])
x2.head()
```

```python
x3 = x2.sort_values(by = "total_acidity", ascending= False)
x3.head()
```

```python
def acidity_ranking(x):
    x = x.query('pH <= 3.2')
    x1 = x.loc[:,['fixed acidity', 'volatile acidity','quality','text_rank']]
    acid_rank = x1.assign(total_acidity=x["fixed acidity"] + x["volatile acidity"]).sort_values(by = "total_acidity", ascending= False)
    return acid_rank  
    
```

```python
print(acidity_ranking(data).head())
```

Question 2: How does sugar alcohol content correlate with wine quality?

1. Group by quality
2. Find average alcohol content
3. Find average sugar 
4. Rank quality by alcohol content

```python
y1 = data.groupby(['text_rank'])
```

```python
y2 = pd.DataFrame()
y3 = y2.assign(meanSugar = y1['residual sugar'].mean(), meanAlcohol= y1['alcohol'].mean())
y3.head()
```

```python
y3 = y3.sort_values(by = 'meanAlcohol', ascending= False)
y3.head()
```

```python
def content(x):
    y1 = x.groupby(['text_rank'])
    y2 = pd.DataFrame()
    y3 = y2.assign(meanSugar = y1['residual sugar'].mean(), meanAlcohol= y1['alcohol'].mean()).sort_values(by = "meanAlcohol", ascending= False)
    return y3
```

```python
print(content(data))
```
